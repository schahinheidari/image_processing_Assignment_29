```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("image/rubix.png")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
R,G,B = cv.split(img)
```

```python
row, col = G.shape
for i in range(row):
    for j in range(col):
        if R[i,j] < 200 and G[i,j] > 50 and B[i,j] > 50:
            R[i,j]=255
            G[i,j]=0
            B[i,j]=0
for i in range(row):
    for j in range(col):
        if R[i,j] > 50 and G[i,j] < 200 and B[i,j] > 50:
            R[i,j]=0
            G[i,j]=255
            B[i,j]=0
for i in range(row):
    for j in range(col):
        if R[i,j] > 50 and G[i,j] > 50 and B[i,j] < 200:
            R[i,j]=0
            G[i,j]=0
            B[i,j]=255
```

```python
img = cv.merge((R , G , B))
plt.imshow(img)
cv.imwrite('output/rubik.jpg',img)
```
